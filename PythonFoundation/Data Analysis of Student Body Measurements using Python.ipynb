### 1.3
Buatlah program untuk menghitung BMI

```python
#program menghitung BMI
def hitung_bmi(bmi):
    if bmi < 18.5 :
        return 'underweight'
    elif bmi < 25:
        return 'normal'
    elif bmi < 30:
        return 'overweight'
    else:
        return 'obesity'
```

### 1.4
Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI

```python
df_19['bmi'] = df_19['Berat Badan (Kg)']/np.power((df_19['Tinggi Badan (Cm)']/100),2)
df_19
```

### 1.5
Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan kolom status BMI

Normal: 18.5-24.9  
Overweight: 25.0-29.9  
Obesity: 30.0-34.9  
Underweight: < 18.5

```python
# Buat kolom baru 'Status BMI' dengan menggunakan fungsi kategori_bmi
df_19['Status BMI'] = df_19['bmi'].apply(hitung_bmi)

# Cetak DataFrame untuk memeriksa hasilnya
df_19
```
